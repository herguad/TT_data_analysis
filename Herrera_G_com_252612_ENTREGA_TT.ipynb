## 1.5 Calidad de datos (nulos y duplicados)

```python
# ============================================
# FUNCIÓN DE CONTROL DE CALIDAD DE DATOS
# ============================================
# Esta función analiza un DataFrame existente (df) y muestra:
# 1️. La cantidad de valores nulos por columna.
# 2️. El total de filas completamente duplicadas.
# 3️. Si se especifica una columna clave, muestra los valores más repetidos de esa columna.

def calidad(df,nombre, clave=None):
  print(f"# {nombre}")
  display(df.isna().sum().to_frame("nulos"))
  dup_filas = df.duplicated(keep=False).sum()
  print("Filas duplicadas: ",dup_filas)
  if clave and clave in df.columns:
    dup_key=df[clave].duplicated(keep=False).sum()
    print(f"Duplicados por columna '{clave}': ", dup_key)
    if dup_key > 0:
      duplicados_ord = (
          df[df[clave].duplicated(keep=False)][clave]
                .value_counts()
                .sort_values(ascending=False)
                )

      print("\n Top 10 valores duplicados más frecuentes: ")
      display(duplicados_ord.head(10))
    else:
      print(f"No se encontraron duplicados en la clave '{clave}'.")
  else:
      if clave:
         print(f"La clave '{clave}' no existe en la DataFrame.")

      else:
        print("No se indicó una clave para analizar duplicados por columna.")
```

## 2.1 Limpieza de Datos
- Eliminación de duplicados
- Normalización de texto en columnas de tipo `object`
- Conversión de fechas a formato datetime (estándar)
- Conversión de `precio` y `cantidad` (df 'Ventas') a tipo numérico.
- Guardado de csv limpios.

```python
# ============================================
# LIMPIEZA Y NORMALIZACIÓN DE LOS DATASETS
# ============================================

# 1️. Crear copias independientes para no modificar los originales

ventas_clean = ventas.copy()
clientes_clean = clientes.copy()
marketing_clean = marketing.copy()

# 2️. Eliminar filas completamente duplicadas

ventas_clean = ventas_clean.drop_duplicates()
clientes_clean = clientes_clean.drop_duplicates()
marketing_clean = marketing_clean.drop_duplicates()
```

```python
# 3. Verificar que las dfs correspondientes quedaron limpias.
calidad(ventas_clean,"Ventas (Limpio)", clave="id_venta")
```

```python
# 4. Normalización de texto en columnas de tipo string
def norm_texto(df):
  for col in df.select_dtypes(include="object").columns:
    df[col] = (
       df[col].astype(str)
       .str.strip()
       .str.replace(r"[\u200b\t\r\n]", "", regex=True)
       .str.replace(" +", " ", regex=True)
       .str.title()
    )
  return df
```

```python
# Aplicar la función para normalizar el texto en las dfs limpias
ventas_clean = norm_texto(ventas_clean)
clientes_clean = norm_texto(clientes_clean)
marketing_clean = norm_texto(marketing_clean)
```

```python
# 5. Normalizar fechas
for df in [ventas_clean, clientes_clean, marketing_clean]:
    for col in df.columns:
        if "fecha" in col.lower():  # detecta columnas con la palabra "fecha"
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
```

```python
# Verificar que la normalización de fechas y texto se aplicó
#-----Revisar tipos------
print(ventas_clean.dtypes)
print(clientes_clean.dtypes)
print(marketing_clean.dtypes)

#------Revisar texto------
print(ventas_clean.head(10))
print(clientes_clean.head(10))
print(marketing_clean.head(10))
```

```python
# 6.1 Normalización de valores numéricos: Campo 'precio'
if "precio" in ventas_clean.columns:
    ventas_clean["precio"] = (
        ventas_clean["precio"]
        .astype(str)                        # Convierte todo a texto
        .str.replace("$", "", regex=False)  # Elimina el símbolo $
        .str.replace(",", "", regex=False)  # Elimina comas de miles
        .str.strip()                        # Quita espacios sobrantes
    )
    ventas_clean["precio"] = pd.to_numeric(ventas_clean["precio"], errors="coerce")
```

```python
# Revisar normalización de columna 'precio'
print(ventas_clean.dtypes)
```

```python
# 6.2 Normalización de valores numéricos: Campo 'cantidad'
if "cantidad" in ventas_clean.columns:
    ventas_clean["cantidad"] = pd.to_numeric(ventas_clean["cantidad"], errors="coerce").astype("Int64")
    # .astype("Int64") usa el tipo entero de pandas que permite valores nulos (NaN)
```

```python
# Revisar normalización de columna 'cantidad'
print(ventas_clean.dtypes)
```

## Reporte post- limpieza de datos

```python
# ============================================
# REPORTE GLOBAL DE CALIDAD DE DATOS
# ============================================
# Esta función lee los dfs limpios
# y muestra un resumen comparativo de nulos, duplicados y tipos.
# ============================================

def reporte_calidad_global(dfs,nombres):
  resumen=[]
  for df,nombre in zip(dfs,nombres):
                       nulos = df.isna().sum().sum() #suma los nulos de todas las columnas
                       duplicados = df.duplicated(keep=False).sum()
                       columnas = len(df.columns)
                       filas = len(df)

                       resumen.append({
                           "Dataset": nombre,
                           "Filas": filas,
                           "Columnas": columnas,
                           "Total nulos": nulos,
                           "Total filas duplicadas": duplicados,
                       })

  reporte = pd.DataFrame(resumen)
  return reporte
```

```python
# Aplicar función "reporte_calidad_global" a las dfs originales y luego a las limpias para comparar
reporte_originales= reporte_calidad_global([ventas, clientes, marketing], ["Ventas Original", "Clientes Original", "Marketing Original"])
reporte_limpias= reporte_calidad_global([ventas_clean, clientes_clean, marketing_clean],["Ventas Limpio", "Clientes Limpio", "Marketing Limpio"])

print(reporte_originales)

print(reporte_limpias)
```

## 2.2 Transformación de Datos

```python
## Transformación de Datos
# 1. Productos de alto rendimiento
# Objetivo:
# - Detectar los productos con mejor desempeño económico (top 20% por ingreso total).
# - Aplicar transformación: calcular ingreso, agregar por producto y filtrar.

# Definir función que busca la primera columna cuyo nombre contenga alguno de los patrones dados
# que devuelva el nombre original de la columna en ese caso
# o None si no encuentra coincidencias

def encontrar_columna(df, patrones):
    for c in df.columns:
        nombre = c.lower()
        if any(p in nombre for p in patrones):
            return c
    return None
```

```python
# Detectar la columna de producto
prod_col = encontrar_columna(ventas_clean, ["producto", "id_producto", "sku", "articulo", "artículo"])
if prod_col is None:
    raise ValueError("No se encontró columna de producto. Renombrá una columna a 'producto' o similar.")
print(prod_col)
```

```python
# Calcular el ingreso por datapoint(registro) multiplicando precio por cantidad.
# Inlcuir resultado en una columna nueva.
# assign(**nuevas_col): crea nuevas columnas y devuelve una copia del df.

ventas_perform = (
    ventas_clean
    .assign(
        ingreso = ventas_clean["precio"] * ventas_clean["cantidad"]
    )
)
```

```python
## Transformación de Datos
#2. Agregación: agrupar el df por una o varias columnas clave y resumir las columnas numéricas por cada grupo.
resumen_prod = (
    ventas_perform.groupby(by=prod_col,dropna=False, as_index=False, observed=False)
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('ingreso', 'size')       # Número de filas en el grupo (cuenta TODO, incluso NaN).
    )
)

#Ver resumen
print(resumen_prod.head(20))
```

```python
# Ordenar resumen_prod por el mayor ingreso_total.
# Redondear precio_promedio a 2 decimales redondeado.
resumen_prod['precio_promedio']=round(resumen_prod['precio_promedio'],2)
resumen_sort=resumen_prod.sort_values(by='ingreso_total',ascending=False)

print(resumen_sort.head(10))
```

```python
# Cálculo del 80% del ingreso total (quantile())
p80_ingreso = resumen_prod["ingreso_total"].quantile(q=0.80,interpolation="linear")

# print(80-ingreso) muestra el límite superior del 80% de los productos.
print(p80_ingreso)
```

```python
# A partir del valor obtenido como límite del 80% del ingreso,
# filtrar productos cuyo ingreso_total está en el 20% superior y
# ordenarlos de mayor a menor por ingreso y unidades.
# en una sola fila

top_ventas = (resumen_prod.query(
        "ingreso_total >= @p80_ingreso",
        engine="python" )
        .sort_values(
        by=["ingreso_total", "unidades"],
        ascending=[False, False],
        na_position="last",
        ignore_index=True
            )
        )

# Mostrar resultados
print(f"Columna de producto detectada: {prod_col}")
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
print("---- Productos de ALTO RENDIMIENTO (top 20% por ingreso) ----")
display(top_ventas.head(20))
```

## 2.3 Agregación de datos
## Resumen de ventas por categoría de producto y análisis de ingresos generados.

```python
## Transformación de datos
# Datasets:
#ventas : dataset original
#ventas_clean: dataset original despues de limpiar datos
#ventas_perform: dataset de ventas_clean con el campo ingreso que es precio*cantidad de cada factura
#resumen_prod: dataset de ventas_perf agrupado por producto y sumado el ingreso de cada venta (ingreso_total)
#top_ventas: el detaset de resumen_prod pero solo incluye los productos mas vendidos (top del 20%)
#ventas_cat: df copia de ventas_perform
#resumen_cat: sobre ventas_cat, agrupo por categoria de producto y sumo ingreso_total

ventas_cat = ventas_perform.copy()

resumen_cat = (ventas_cat.groupby(by='categoria', dropna=False,as_index=False)
             .agg(ingreso_total=('ingreso', 'sum'),
                  unidades=('cantidad', 'sum'),
                  ventas=('ingreso', 'size'),
                  precio_promedio=('precio', 'mean'))
             .sort_values(
                 by='ingreso_total',
                 ascending=False,
                 na_position='last',
                 ignore_index=True # Reindexar desde 0
             ))

# Nueva columna para incluir ticket promedio por venta.
resumen_cat = resumen_cat.assign(
    ticket_promedio_por_venta = resumen_cat['ingreso_total'] / resumen_cat['ventas'])

print("Resumen por categoría (ordenado por ingreso_total):")
display(resumen_cat.head(20))
```

## Cálculo del rango intercuartílico (IQR)

```python
q1 = df1['ingreso_total'].quantile(0.25)
q2 = df1['ingreso_total'].quantile(0.5)
q3 = df1['ingreso_total'].quantile(0.75)
iqr = q3 - q1

print("Q1:", round(q1,3))
print("Mediana (Q2):", round(q2,3))
print("Q3:", round(q3,3))
print("IQR:", round(iqr,3))
```

## Detección de outliers usando el IQR

```python
lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr

outliers = df1[(df1['ingreso_total'] < lim_inf) | (df1['ingreso_total'] > lim_sup)]

print("Límite inferior:", lim_inf)
print("Límite superior:", lim_sup)
print("\nOutliers detectados:")
print(outliers)
```

```python
# Grafico de líneas que muestra los ingresos diarios y los ingresos acumulados

import matplotlib.pyplot as plt
import pandas as pd

# Convertir ventas_perform['fecha_venta'] a tipo datetime
ventas_perform['fecha_venta'] = pd.to_datetime(ventas_perform['fecha_venta'], errors='coerce')

#Agrupar ventas por fecha y sumar los ingresos por dia
ingreso_diario=ventas_perform.groupby('fecha_venta')['ingreso'].sum().sort_index()

#Calcular ingreso acumulado
ingresos_acum = ingreso_diario.cumsum()

# Graficar evolución de ventas por fecha
plt.figure(figsize=(12, 6))

# Línea 1: ingresos diarios
plt.plot(ingreso_diario, color=color_teal, label='Ingreso Diario')

# Línea 2: ingresos acumulados
plt.plot(ingresos_acum, color=color_indigo, label='Ingreso Acumulado', linewidth=2)

# Personalización del gráfico
plt.title('Evolución de Ingresos por fecha de venta')
plt.xlabel('Fecha de venta')
plt.ylabel('Ingresos ($)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Interpretación:
# Si aparecen saltos abruptos → picos de venta (promociones o campañas de marketing).
```
